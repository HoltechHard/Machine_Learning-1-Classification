## DEFINIÇÃO 

Modelo computacional que adota a propriedade de pequenas unidades de computação para um processamento paralelo   
e não centralizado, na qual os pesos entre as unidades são o principal meio de armazenamento de informação

## ARQUITETURA DE UMA REDE MLP 

Está definida por:  
1) número de neurônios  
2) número de camadas  
3) tipos de conexão entre as camadas
